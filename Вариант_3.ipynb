1. Необходимо построить графики. Стиль оформления графиков (цвета, границы, толщину линий, подписи и т.п.) должен соответствовать представленному ниже. Границы значений оси абсцисс должны быть равны минимальному и максимальному значению отображаемого столбца. Функция для построения графика представлена на оси Y.

ps. Не выполняйте ячейки с графиками чтоб не удалить пример.



```python
import matplotlib.pyplot as plt
import math

x = [i/100 for i in range(-800, 800)]
y = [(x**3)-(x**4) + (25*math.sin(5*x)) + 3 for x in x]
plt.title('График функции')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('$x^3-x^4+25\sin(5x)+3$')
plt.plot(x,y,color='green')
```

```python
x = [0, 5, 10, 15, 20]
y = [50, 40, 45, 60, 55]
plt.figure(figsize=(12, 7))

plt.subplot(2, 2, 1)
plt.plot(x, y,)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y, linestyle='dashed')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y, linestyle='dashdot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y, linestyle='dotted')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
